```python
import csv
import matplotlib.pyplot as plt
```

```python
f = open('age_eng.csv')
data = csv.reader(f)
x = []; y = [] ; z = []
labels = []
n = 0
for row in data:
    if row[0]!='행정구역':
        labels.append(row[0])
        y.append(float(row[1]))
        z.append(float(row[2]))
        n+=1
        x.append(n)
        
plt.title("age graph")
plt.grid(True)
plt.xlim(1,n)
plt.xticks(x,labels,rotation = 'vertical')
plt.plot(x,y,'r:',label = 'male')
plt.plot(x,z,'b--',label = 'femael')
plt.legend(loc = 2)
plt.show()
```
